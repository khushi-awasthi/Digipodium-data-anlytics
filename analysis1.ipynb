```python
marks['Total'].plot(kind ='pie',figsize=(10,5),title='Total marks')
```

```python
#Sorting
marks.sort_values(by='Total',ascending=False,inplace=True)
```

```python
marks
```

```python
marks
```

```python
marks['percentage']=marks['Total']/5
```

```python
marks.sort_values(by=['Total','percentage'],ascending=False,inplace=True)
```

```python
marks
```

```python
marks['percentage'].plot(kind ='pie',figsize=(10,5),title='percentage')
```

```python
fig,ax=plt.subplots(figsize=(10,5),ncols=2)
marks['Total'].plot(kind='bar',ax=ax[0])
marks['percentage'].plot(kind='kde',ax=ax[1])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,5),ncols=2)
marks['Total'].plot(kind='bar',ax=ax[0])
cols=['DBMS','DM','ECO','GFX','SE']
marks[cols].plot(kind='kde',ax=ax[1])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,5),ncols=2)
marks['Total'].plot(kind='hist',ax=ax[0])
cols=['DBMS','DM','ECO','GFX','SE']
marks[cols].plot(kind='hist',ax=ax[1])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,5),ncols=2)
marks['Total'].plot(kind='line',ax=ax[0])
cols=['DBMS','DM','ECO','GFX','SE']
marks[cols].plot(kind='line',ax=ax[1])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,5),ncols=2)
marks['Total'].plot(kind='area',ax=ax[0])
cols=['DBMS','DM','ECO','GFX','SE']
marks[cols].plot(kind='area',ax=ax[1])
plt.show()
```
