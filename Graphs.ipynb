```python
import numpy as np
import matplotlib.pyplot as plt
```

get the data /visualise the data

```python
rainfall = [67,20,10,50,26,59]
months = ['jan','feb','Mar','Apr','may','Jun']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')#theme
```

Visualise the data

```python
plt.bar(months,rainfall,color='#f90129')
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    explode=(0,0,0,0,0,0.1)
    
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    
    
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.d',
    
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%.1f%%',
    wedgeprops={'width':.4},
    pctdistance=.8
    
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%.1f%%',
    wedgeprops={'width':.4,'edgecolor':'#6666ff','linewidth':2},
    pctdistance=.8
    
)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
            )
plt.show()
```
